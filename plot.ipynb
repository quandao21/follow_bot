```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

run_time = np.arange(0,20)
# Read the CSV file into a DataFrame
df_no_opt = pd.read_csv('not_opt.csv')
df_opt = pd.read_csv('opt.csv')
print(df_no_opt['ransactime'])
# Create a line plot of the 'threshold' column against the 'precision' column

plt.plot( run_time,df_no_opt['ransactime'],linestyle='-.', linewidth=2 , color ="orange" )
plt.plot( run_time,df_opt['ransactime'],linestyle='-.', linewidth=2 , color ="red" )

# Set the title and axis labels
plt.title('Accuracy vs. Threshold')
plt.xlabel('Threshold')

plt.ylabel('Accuracy')

# Display the plot
plt.show()
```
